```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(10,100,100)
```

```python
def hypothesis(theta,X):
    return X*theta[1] + theta[0]
```

```python
noise = np.random.randint(-7,7, len(X))
```

```python
theta = [1,1]
y = hypothesis(theta, X) + noise
```

```python
plt.scatter(X,y)
```

```python
plt.figure()
plt.scatter(X,y)
plt.ylim(0,300)
plt.show()
```

```python
class LinearRegression():
    def __init__(self, lr = .01, iters = 1):
        self.lr = lr
        self.iters = iters
    def fit(self, X, y):
        self.X = X
        self.y = y
        self.theta = np.zeros([2], dtype = float)
        
        for i in range(self.iters):
            self.gradient_desc()
            
    def gradient_desc(self):
        yh = LinearRegression.hypothesis(self.theta, self.X)
        change_slope = np.sum((yh-self.y)*X)/len(X)  #divided by len(X) so that we get kinda avg value
                                            # ^^^
        change_inter = np.sum((yh-self.y))/ len(X)#Change in intercept
        change_slope = change_slope*self.lr
        change_inter = change_inter*self.lr
        
        self.theta[0] -= change_inter
        self.theta[1] -= change_slope
    
    def predict(self,X):
        return LinearRegression.hypothesis(self.theta,X)
    
    @classmethod
    def hypothesis(cls, theta, X):
        return X*theta[1]+theta[0]
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

# Lin Reg by sklearn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression?
from sklearn.linear_model import LinearRegression
```

```python
X, y = make_regression(100, n_features=1, bias = 100, noise = 20)
plt.scatter(X,y)
```

```python
model = LinearRegression()
model.fit(X,y)
yh = model.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,yh, color = 'red')
```

```python
model.score(X,y)
```
